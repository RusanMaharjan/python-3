# Nested IF

```python
# Write a program to check whether a person is eligible to vote or not.
# condition should be person's age must be more than equals 18 and must be a Nepali citizen.
```

```python
age = int(input("Enter age: "))
citizen = input("Enter citizen details: ")

if age >= 18: # True
    if citizen == 'Nepali': # False
        print("A person is eligible to vote!!")
    else:
        print("A person is not a Nepali citizen.")
else:
    print("A person age must be more than or equals to 18")
```

```python
'''
a
b
choice

1. add = a + b
2. Subtract
3. multiply
4. Divide
'''
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

choice = input("Enter value among +, -, /, *")

match choice:
    case '+':
        add = a + b
        print("Addition of a and b is:",add)
    case '-':
        sub = a - b
        print("Subtraction of a and b is:",sub)
    case '*':
        mul = a * b
        print("Multiplication of a and b is:",mul)
    case '/':
        div = a / b
        print("Division of a and b is:",div)
    case _:
        print("Invalid Value")
```
