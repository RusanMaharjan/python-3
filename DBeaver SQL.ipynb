```python
# package for connecting python with database
import sqlite3
```

```python
# Create database connection
conn = sqlite3.connect('Sales.db')
```

```python
# does interaction between Python and database
cursor = conn.cursor()
```

## create products table

```python
try:
    cursor.execute("""
        Create Table Products(
            Product_Id int Primary Key,
            Product_Name text,
            Product_Price int,
            Product_Description text
        )
    """)
    print('Product Table Created.')
except:
    print('Table Product Already Exists.')
```

## Create Orders Table

```python
try:
    cursor.execute("""
        Create table Orders(
            Order_id int primary key,
            order_location text,
            order_quantity int,
            order_total_price int
        )
    """)
    print('Order Table Created.')
except:
    print('Order Table Already Created.')
```

# fetch all data

```python
cursor.execute("""
    select * from products
""")
data = cursor.fetchall()
for d in data:
    print(d)
```

# Insert data

```python
cursor.execute("""
    Insert into products
    (product_id, product_name, product_price, product_description)
    values
    (1, 'TV', 200000, 'Smart TV')
""")
print('Data inserted to product table.')
conn.commit()
```

# Update Data

```python
cursor.execute("""
    update products set product_price = 30
    where product_id = 5
""")
print('Product Data Updated Successfully.')
conn.commit()
```

# Where Clause

```python
'''
SQL Where Clause -> Used for data filtering
comparison operator
logical operator

Like -> Pattern Match
Between -> Alternative of AND Operator
IN -> Alternative of OR Operator

'''
```

```python
cursor.execute("""
    select * from products where product_price < 200000
""")
data = cursor.fetchall()
for d in data:
    print(d)
```

```python
cursor.execute("""
    select * from products 
    where product_price > 20 and product_price < 200000
""")
data = cursor.fetchall()
for d in data:
    print(d)
```
