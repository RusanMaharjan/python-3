```python
def add(a, b):
    return a + b

def dif(a, b):
    return a - b

def mul(a, b):
    return a * b

def div(a, b):
    return a / b
```

```python
def calculator(a, b):
    addition = add(a, b)
    subtraction = dif(a, b)
    multiplication = mul(a, b)
    division = div(a, b)
    return addition, subtraction, multiplication, division
```

```python
calculator(4, 2)
```
