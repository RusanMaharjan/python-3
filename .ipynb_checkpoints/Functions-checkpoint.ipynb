```python
def calculation():

    a = int(input("Enter a number: "))
    b = int(input("Enter second number: "))
    choice = int(input("Enter number from 1 to 4: "))

    if choice == 1:
        add = a + b
        print(add)
    elif choice == 2:
        sub = a - b
        print(sub)
    elif choice == 3:
        mul = a * b
        print(mul)
    elif choice == 4:
        div = a / b
        print(div)
    else:
        print("Invalid Number!!")
```

```python
calculation()
```

```python
'''
Function it should accept two arguments: student roll number and full name
Name of student: Name
Roll Number of student: roll number

atleast 5 subject marks

Total Marks
Average Marks

Grade: A, B, C, D, E, NG

'''
```

```python
def Student(full_name, roll_number):
    print("Student Name:",full_name)
    print("Roll Number:",roll_number)

    marks1 = int(input("Enter marks for subject 1: "))
    marks2 = int(input("Enter marks for subject 2: "))
    marks3 = int(input("Enter marks for subject 3: "))
    marks4 = int(input("Enter marks for subject 4: "))
    marks5 = int(input("Enter marks for subject 5: "))

    total_marks = marks1 + marks2 + marks3 + marks4 + marks5

    avg_mark = total_marks/5

    if avg_mark >=80 and avg_mark <= 100:
        print("Grade: A")
    elif avg_mark >=60 and avg_mark < 80:
        print("Grade: B")
    elif avg_mark >=50 and avg_mark < 60:
        print("Grade: C")
    elif avg_mark >=40 and avg_mark < 50:
        print("Grade: D")
    else:
        print("Grade: NG")
```

```python
Student("Ram", 2)
```

```python
Student()
```

# Default Argument and Keyword Argument

## Default Argument

```python
def person(name="Unknown Name", age=0, address='Kathmandu'):
    print("Name:",name)
    print("Age:",age)
    print("Address:",address)
```

```python
person("Ram", 30, "Lalitpur")
```

```python
person(age=20)
```

## Keyword Argument

```python
def person1(name, age, address):
    print("Name:",name)
    print("Age:",age)
    print("Address:",address)
```

```python
person1(name="ABC", age=20, address='Lalitpur')
```

# (All Arguments)*Args , (All Keyword Arguments)**kwargs

```python
def phone(phone_name, *phone_details):
    print("Phone Name:",phone_name)
    for each_arg in phone_details:
        print(each_arg)
    print("All Arguments:",phone_details)
```

```python
phone("Samsung", "S25", 2022)
```

```python
def phone_details(phone_name, **kwargs):
    print("Phone Name:",phone_name)
    print(kwargs)
    for key, value in kwargs.items():
        print(f"{key.capitalize()}: {value}")
```

```python
phone_details("Samsung", model='s24', build_year='2022', os='Android 14')
```

# Lambda function

```python
# add value using lambda function
add = lambda x: x + 10
```

```python
add(5)
```

```python
add = lambda x, y: x + y
```

```python
add(10, 5)
```

```python
def sum_all(*args):
    total = 0
    for i in args:
        total += i
    print(total)
```

```python
sum_all(1, 2, 3)
```
