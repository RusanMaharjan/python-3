# Install DBeaver for SQL Task in Python

```python
'''
SQL (Structured Query Language) / Database
- SQLite
- Create Database
- Create Table
- Insert Data
- Update Data
- Delete Data
- Data Filtering (Where clause)
'''
```

```python
# package for connecting python with database
import sqlite3
```

```python
# Create database connection
conn = sqlite3.connect('Sales.db')
```

```python
# does interaction between Python and database
cursor = conn.cursor()
```

## create products table

```python
try:
    cursor.execute("""
        Create Table Products(
            Product_Id int Primary Key,
            Product_Name text,
            Product_Price int,
            Product_Description text
        )
    """)
    print('Product Table Created.')
except:
    print('Table Product Already Exists.')
```

## Create Orders Table

```python
try:
    cursor.execute("""
        Create table Orders(
            Order_id int primary key,
            order_location text,
            order_quantity int,
            order_total_price int
        )
    """)
    print('Order Table Created.')
except:
    print('Order Table Already Created.')
```
