# Break and Continue

```python
'''
Condition True:
    break (program stop)

condition True:
    continue (value skip)
'''
```

```python
# find numbers from 1 to 10 but program needs to be stopped when it finds number 5
for i in range(1, 11):# 5
    if i == 5: # 5 == 5 # True
        print("Program stopped in number 5.")
        break
    print(i)
```

```python
# find numbers from 1 to 10 but program needs to be continued when it finds number 5
for i in range(1, 11):
    if i == 5:
        continue
    print(i)
```

```python
# user input 5 times if user inputs 0 value then program needs to be stopeed.
for i in range(1, 6): # 1, 2, 3, 4, 5
    n = int(input("Enter any number: "))
    if n == 5:
        print("You have entered value 5. Program Stopped!!")
        break # Forcefully stop program
    print("You have entered:",n)
```

```python
for i in range(5): # 0, 1, 2, 3, 4
    n = int(input("Enter any number: "))
    if n == 5:
        print("You have entered value 5. Program Stopped!!")
        break # Forcefully stop program
    print("You have entered:",n)
```

```python
# 1. Snaks
# 2. Lunch
# 3. Dinner
# 4. Breakfast
# 5. Exit
# 1-4 -> You have orderd snaks
# 5 -> Thank you for ordering food.
for i in range(5):
    choice = int(input("Enter a number: "))
    if choice == 1:
        print("snacks")
    elif choice == 2:
        print("Lunch")
    elif choice == 3:
        print("Dinner")
    elif choice == 4:
        print("Breakfast")
    elif choice == 5:
        print("Thank you for ordering food.")
        break
    else:
        print("Invalid Number.")
else:
    print("You have orderd 5 items.")
```

# While Loop

# Guess the Number game

```python
secret_value = 7
guess = 0
while guess != secret_value:
    guess = int(input("Enter guess value: "))
    if guess == secret_value:
        print("You have guessed correct value!!")
    else:
        print("Incorrect value. Try again.")
```

```python
'''
Create a food order program: 
Where user can order any items, they want to order 
Loop should run until user choose number 5 to exit. 
1. Momo 
2. Chowmin 
3. Pizza 
4. Drinks 
5. Exit
'''
```

```python
choice = 0
while choice != 5:
    print('1. Momo')
    print('2. Chowmin')
    print('3. Pizza')
    print('4. Drinks')
    print('5. Exit')
    choice = int(input("Enter number from 1 - 5: "))

    if choice == 1:
        print("You have ordered Momo.")
    elif choice == 2:
        print("You have ordered Chowmin.")
    elif choice == 3:
        print("You have ordered Pizza.")
    elif choice == 4:
        print("You have ordered Drinks.")
    elif choice == 5:
        print("Thank you for choosing us!!")
    else:
        print("Invalid Input!!")
```

```python
pin = 1234
attempt = 3

# pin value matched with actual pin then print Access Granted
# if value not matched then print invalid pin number. Try again
correct_pin = 0
while correct_pin != pin:
    print("Remaining attempt:",attempt)
    correct_pin = int(input("Enter your pin: "))

    if correct_pin == pin:
        print("Access Granted!!")
    else:
        attempt = attempt - 1
        print("Invalid Pin!!")

    if attempt == 0:
        print("Card Blocked!!")
        break
```

```python
pin = 1234
attempt = 3

# pin value matched with actual pin then print Access Granted
# if value not matched then print invalid pin number. Try again
correct_pin = 0
while correct_pin != pin:
    correct_pin = int(input("Enter your pin: "))

    if correct_pin == pin:
        print("Access Granted!!")
    elif attempt == 0:
        print("Card Blocked!!")
        break
    else:
        attempt = attempt - 1
        print("Invalid Pin!!")
        print("Remaining Attempt:",attempt)
```
