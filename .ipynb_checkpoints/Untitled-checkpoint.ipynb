```python
# Find the greater number among two numbers
```

```python
num1 = 5
num2 = 6
num3 = 10
num4 = 11
```

```python
if num1 > num2: # False
    print("Num1 is greater than num2")
else:
    print("Num2 is greater than num1")
```

```python
if num1 > num2: # False
    print("Num1 is greater than num2.")
elif num3 < num4:
    print("Num3 is lesser than num4.")
elif num4 < num1:
    print("Num4 is lesser than num1.")
else:
    print("No any condition matched.")
```

```python
# take input of 5 subjects, total marks, average marks, grade
```

```python
sub1 = 75
sub2 = 80
sub3 = 45
sub4 = 75
sub5 = 75
```

```python
total_marks = sub1 + sub2 + sub3 + sub4 + sub5
print(total_marks)
```

```python
average_marks = total_marks/5
print(average_marks)
```

```python
if average_marks > 80:
    print("A")
elif average_marks < 80 and average_marks > 50:
    print("B")
else:
    print("C")
```
