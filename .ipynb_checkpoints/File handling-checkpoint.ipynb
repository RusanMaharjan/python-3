# File Handling

```python
'''
File Handling Operations
---------------------------
r -> read
w -> write
a -> append
rb -> read binary
wb -> write binary
'''
```

```python
'''
text file
CSV file
Dictionary File
JSON File
Binary file copy and paste (image, pdf, word, excel)
'''
```

# Text File

## Creating text file

```python
# open(file_name, operation)
file = open('text_file.txt', 'w')
content = file.write('Hello World!!')
file.close()
```

## Read text file

```python
file = open('text_file.txt', 'r')
content = file.read()
print(content)
```

## Append values in text file

```python
file = open('text_file.txt', 'a')
content = file.write('\n New Hello World!!')
file.close()
```

# CSV File (Comma Seperated Value)

```python
import csv
```

```python
countries_destination = [
    ['Countries', 'Destination', 'Price'],
    ['USA', 'New York', '$5000'],
    ['Italy', 'Rome', '$3000'],
    ['Nepal', 'Mustang', '$1000']
]
```

```python
# open(file_name, operation, newline)
with open('country.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(countries_destination)
```

## read csv file

```python
with open('country.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
# Dictionary
personal_details = [
    {
        "Name": "John",
        "Age": 30,
        "Address": "New York"
    },
    {
        "Name": "Charles",
        "Age": 20,
        "Address": "Canada"
    },
    {
        "Name": "Dutch",
        "Age": 35,
        "Address": "Ottawa"
    }
]
```

## Write dictionary value in csv file

```python
with open('personal_details.csv', 'w', newline='') as file:
    columns = ['Name', 'Age', 'Address']
    writer = csv.DictWriter(file, fieldnames=columns)
    writer.writeheader()
    writer.writerows(personal_details)
```

```python
with open('personal_details.csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(row)
```

# Json File

```python
personal_details = [
    {
        "Name": "John",
        "Age": 30,
        "Address": "New York"
    },
    {
        "Name": "Charles",
        "Age": 20,
        "Address": "Canada"
    },
    {
        "Name": "Dutch",
        "Age": 35,
        "Address": "Ottawa"
    }
]
```

```python
import json
```

```python
# Write -> dump()
# Read -> load()
```

```python
with open('personal_details.json', 'w') as file:
    json.dump(personal_details, file)
```

```python
with open('personal_details.json', 'r') as file:
    data = json.load(file)
    for each_data in data:
        print(each_data)
```
