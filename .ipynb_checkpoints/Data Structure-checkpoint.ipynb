```python
# What is dictionary?
# what is key-value pair?
```

```python
choice = 0
order_list = []
while choice != 5:
    print('1. Momo')
    print('2. Chowmin')
    print('3. Pizza')
    print('4. Drinks')
    print('5. Exit')
    choice = int(input("Enter number from 1 - 5: "))

    if choice == 1:
        msg = "You have ordered Momo."
        print(msg)
        order_list.append(msg)
    elif choice == 2:
        msg = "You have ordered Chowmin."
        print(msg)
        order_list.append(msg)
    elif choice == 3:
        msg = "You have ordered Pizza."
        print(msg)
        order_list.append(msg)
    elif choice == 4:
        msg = "You have ordered Drinks."
        print(msg)
        order_list.append(msg)
    elif choice == 5:
        print("Thank you for choosing us!!")
    else:
        print("Invalid Input!!")
print()
print("Your final order list:")
for i, value in enumerate(order_list):
    print(f"{i+1}: {value}")
```

# Dictionary

```python
'''
Dictionary = {
    'key': 'value'
}

It is also mutable data structure.

Add, Update, Delete
'''
```

```python
empty_dictionary = {}
type(empty_dictionary)
```

```python
details = {
    'Name': 'Alex',
    'Age': 30,
    'Address': 'London'
}
details
```

```python
# details[0]
details['Name']
```

```python
details1 = {
    'Name': ['Alex', 'Bob'],
    'Age': [30, 50],
    'Address': ['London', 'New York']
}
details1
```

```python
details1['Name']
```

```python
for i in details1:
    print(i)
```

# Loop in dictionary

```python
'''
.keys(), # extract dictionary keys
.values(), # extract dictionary values
.items() # extract both keys and values
'''
# Keys
for key in details1.keys():
    print(key)

print()

# Values
for value in details1.values():
    print(value)

print()

# Items
for key, value in details1.items():
    print(f"{key}: {value}")
```
