```python
try:
    num1 = int(input('Enter a number: '))
    num2 = int(input('Enter a number: '))

    div = num1/num2
    print(div)
except ZeroDivisionError:
    print('Error: Number cannot be divisible by Zero.')
except ValueError:
    print('Error: Only number values can be accepted by input field.')
```

```python
try:
    file_name = input('Enter file name here: ')
    file = open(file_name, 'r')
except FileNotFoundError:
    print(f'File name {file_name} not found in directory.')
else:
    content = file.read()
    print(content)
    file.close()
finally:
    print('Task Successfull!!')
```

```python
# Check whether a persons age is more than 18 or not to vote.
class AgeError(Exception):
    pass
```

```python
def check_age(age):
    if age > 18:
        print('You are eligible to vote!!')
    else:
        raise AgeError('AgeError: Age must be more than 18 to vote')
```

```python
try:
    age = int(input('Enter your age: '))
    check_age(age)
except AgeError as e:
    print(e)
```

```python
class IndexOutOfBoundError(Exception):
    pass
```

```python
def index(data_list, target_index):
    if target_index < 0 or target_index >= len(data_list):
        raise IndexOutOfBoundError(f'Index {target_index} is out of bound of list {len(data_list)}')
    return data_list[target_index]
```

```python
try:
    my_list = [1, 2, 3, 4]
    result = index(my_list, 3)
    print(result)
except IndexOutOfBoundError as e:
    print(e)
```
