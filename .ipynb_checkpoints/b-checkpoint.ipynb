```python
import random
```

```python
class Bank:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance
        self.account_number = ''.join([str(random.randint(1, 9)) for i in range(16)])

        print(f'Account holder {self.account_holder} created with A/C no. {self.account_number}')

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f'Rs. {amount} deposited to A/C no. {self.account_number}.')
        else:
            print(f'Amount must be more than 0.')

    def withdraw(self, amount):
        if amount < self.balance:
            self.balance -= amount
            print(f'Rs. {amount} withdrawn from A/C no. {self.account_number}.')
        else:
            print(f'Amount must be less than balance.')

    def print_details(self):
        print(f'Account Holder: {self.account_holder}')
        print(f'Account Number: {self.account_number}')
        print(f'Account Balance: {self.balance}')
```

```python
all_account = []

def findByAccountNumber(acc_number):
    for account in all_account:
        if account.account_number == acc_number:
            return account
    return None
```

```python
while True:
    print(f'Choose what you want to do?')
    print(f'1. Create New Account')
    print(f'2. Deposit Amount')
    print(f'3. Withdraw Amount')
    print(f'4. Display Account Details')
    print(f'5. Exit')

    choice = input('Enter your choice number: ')

    if choice == '1':
        name = input('Enter account holder name: ')
        dep_balance = int(input('Enter initial balance to deposit: '))

        new_acc = Bank(name, dep_balance)
        all_account.append(new_acc)
        print('Account Created')
    elif choice in ['2', '3', '4']:
        acc_number = input('Enter account number: ')
        find_acc = findByAccountNumber(acc_number)
        if choice == '2':
            if find_acc:
                balance = int(input('Enter amount to deposit: '))
                find_acc.deposit(balance)
            else:
                print('erro')
        elif choice == '3':
            if find_acc:
                balance = int(input('Enter amount to withdraw: '))
                find_acc.withdraw(balance)
            else:
                print('error')
        elif choice == '4':
            if find_acc:
                find_acc.print_details()
            else:
                print('err')
    elif choice == '5':
        print('Thank you for choosing us.')
        break
    else:
        print('Invalid Value.')
```
