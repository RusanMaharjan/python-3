```python
# parent class
class Person:
    # constructor -> 3 parameter
    def __init__(self, name, age, address):
        # instance_Variable = local_variable
        self.name = name
        self.age = age
        self.address = address
```

```python
# Child class
class Staff(Person):
    def __init__(self, name, age, address, position, post, staff_type):
        super().__init__(name, age, address)
        self.position = position
        self.post = post
        self.staff_type = staff_type

    def print_details(self):
        print(self.position)
        print(self.post)
        print(self.staff_type)
        print(self.name)
        print(self.age)
        print(self.address)
```

```python
p = Staff("Ram", 20, "Kathmandu", "Data Analyst", 'Senior', "Full Time")
```

```python
p.print_details()
```

```python
'''
Vehicle
- constructor(name, color, weight)
- Function start -> Vehicle Started.
- Function stop -> Vehicle Stopped.

Two-Wheelers extends vehicle
- constructor(brand, speed)
- Print function (all details)
- Print start and stop function.
'''
```

```python
class Vehicle:
    def __init__(self, name, color, weight):
        self.name = name
        self.color = color
        self.weight = weight

    def start(self):
        return "Vehicle Started."

    def stop(self):
        return "Vehicle Stopped."
```

```python
class TwoWheelers(Vehicle):
    def __init__(self, name, color, weight, brand, speed):
        super().__init__(name, color, weight)
        self.brand = brand
        self.speed = speed
        
    def print_details(self):
        print(f"Name of vehicle: {self.name}")
        print(f"Color of vehicle: {self.color}")
        print(f"Weight of vehicle: {self.weight}")
        print(f"Brand of vehicle: {self.brand}")
        print(f"Speed of vehicle: {self.speed}")
        print(self.start())
        print(self.stop())
```

```python
tw = TwoWheelers("AB", "Red", "152kg", "DE", 200)
```

```python
tw.print_details()
```

```python
class Staff(Person):
    def __init__(self, name, age, address, job, position, staff_type):
        super().__init__(name, age, address)
        self.job = job
        self.position = position
        self.staff_type = staff_type
```

```python
class Company(Staff):
    def __init__(self, name, age, address, job, position, staff_type, 
                 company_name, company_address):
        super().__init__(name, age, address, job, position, staff_type)
        self.company_name = company_name
        self.company_address = company_address
        
    def print_details(self):
        print(self.staff_type)
        print(self.name)
        print(self.age)
        print(self.address)
        print(self.position)
        print(self.job)
        print(self.company_name)
        print(self.company_address)
```

```python
c.print_details()
```

# Encapsulation

```python
'''
Data Hiding
------------------
access modifier
1. Private (__)
2. Public
3. Protected (_)

it is used with instance variable.
'''
```

```python
class Person:
    def __init__(self, name, age, address):
        # Private Modifier
        self.__name = name
        self.__age = age
        self.__address = address

    def print_details(self):
        print("Name:",self.__name)
        print("Age:",self.__age)
        print("Address:",self.__address)
```

```python
class Staff(Person):
    def __init__(self, name, age, address, job, position):
        super().__init__(name, age, address)
        self.__job = job
        self.__position = position

    def staff_details(self):
        print("Job:",self.__job)
        print("Position:",self.__position)
```

```python
s = Staff("Ram", 30, "Lalitpur", "Data Analyst", "Senior")
```

```python
s.print_details()
```

```python
s.staff_details()
```

# Abstraction

```python
'''
Data Hiding
---------------
Only show important information to user.
If we created Abstract Class at least one Abstract method should be created.
Interface -> Class without body. Methods/Functions
'''
```

```python
# Abstract Base Class
from abc import ABC, abstractmethod
```

```python
# Abstract Base Class
class Payment(ABC):
    @abstractmethod
    def pay(self):
        pass

    @abstractmethod
    def connection(self):
        pass
```

```python
class ESewa(Payment):
    def pay(self, amount):
        print(f"{amount} paid through ESewa.")

    def connection(self):
        print("connection successfull!!")
```

```python
e = ESewa()
```

```python
e.pay(300)
```

```python
e.connection()
```
